
# Attribute Information

1) gender: "Male", "Female" or "Other"

2) age: age of the patient

3) hypertension: 0 if the patient doesn't have hypertension, 1 if the patient has hypertension

4) heart disease: 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease 5) Ever-married: "No" or "Yes"

6) work type: "children", "Govtjov", "Never worked", "Private" or "Self-employed"

7) Residencetype: "Rural" or "Urban"

8) avg glucose level: average glucose level in blood

9) BMI: body mass index

10) smoking_status: "formerly smoked", "never smoked", "smokes" or "Unknown"

11) stroke: 1 if the patient had a **stroke** or 0 if not

```python
"""Study this Code"""

# Clean style with no grid
sns.set_style("white")

# Set figure size
plt.figure(figsize=(8, 6))

# Create the histogram
ax = sns.histplot(data=df, x="gender",
                  shrink=0.8,
                  edgecolor='blue',
                  linewidth=1.2)

# Add count labels on top of each bar
for patch in ax.patches:
    height = patch.get_height()
    ax.text(patch.get_x() + patch.get_width() / 2,
            height + 1,  # slightly above the bar
            int(height),
            ha='center', va='bottom', fontsize=12, fontweight='bold')

# Add title and axis labels
plt.title("Number of female and male", fontsize=16, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("")  # Remove y-axis label

# Remove y-axis ticks and lines
ax.tick_params(axis='y', left=False, labelleft=False)
sns.despine(left=True)  # Remove left spine

# Show the plot
plt.tight_layout()
plt.show()
```

```python
""" Read This before """



# ------------------------------------------
# Step 2: Encode Categorical Variables
# ------------------------------------------

# Correlation only works with numerical data.
# So if you have categorical variables like 'gender', 'work_type', etc.,
# you need to convert them into numerical format first.

# Example: Encoding 'gender' column where 'Female' becomes 1 and 'Male' becomes 0
df2['gender'] = df2['gender'].replace({'Female': 1, 'Male': 0})

# Check unique values in 'work_type' to understand the categories
print(df2['work_type'].unique())

# Option 1: Use replace() to manually assign numerical values to each category in 'work_type'
df2['work_type'] = df2['work_type'].replace({
    'Private': 0,
    'Self-employed': 1,
    'Govt_job': 2,
    'Children': 3,
    'Never_worked': 4
})

# Option 2 (Recommended for models): Use One-Hot Encoding instead of manually assigning numbers
# Uncomment the following line if you prefer one-hot encoding
# df2 = pd.get_dummies(df2, columns=['work_type'], drop_first=True)

# ------------------------------------------
# Step 3: Build Correlation Matrix
# ------------------------------------------

# Now that all categorical variables are encoded as numbers, we can safely compute correlation
correlation_matrix = df2.corr()

# Display the correlation matrix
print(correlation_matrix)

# ------------------------------------------
# Step 4: Optional - Visualize Correlation Matrix
# ------------------------------------------

# It's often helpful to visualize the correlation using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size for better readability
plt.figure(figsize=(10, 8))

# Plot the heatmap with correlation values
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title to the heatmap
plt.title('Correlation Matrix')

# Display the plot
plt.show()
```
